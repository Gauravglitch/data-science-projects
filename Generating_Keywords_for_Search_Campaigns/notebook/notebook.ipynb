```python
import pandas as pd
```

```python
#Pair words with products
word_list = ['buy','price','discount','promotion','promo','shop']
product_list = ['sofas','convertible sofas','love seats','recliners','sofa beds']
keyword_list =[]
for word in word_list:
    for product in product_list:
        keyword_list.append([product,word + " " + product])
        keyword_list.append([product,product +" " + word])
print(keyword_list)
```

```python
#Convert the list into a DataFrame with descriptive column names
keywords_df = pd.DataFrame(data=keyword_list, columns=['Ad Group','Keyword'])
print(keywords_df)
```

```python
#Add a campaign and match type column
keywords_df['Campaign'] = 'SEM_Sofas'
keywords_df['Criterion Type'] = 'Exact'
print(keywords_df)
```

```python
#Save a DataFrame as a CSV file
keywords_df.to_csv("keywords.csv")
```
