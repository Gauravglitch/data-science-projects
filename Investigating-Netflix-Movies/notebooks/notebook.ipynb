```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
#Filtered the data for movies released in the 1990s
type_movie = netflix_df[netflix_df['type']=='Movie']
released_from_1990= type_movie[type_movie['release_year'] >=1990]
released_up_to_1999 = released_from_1990[released_from_1990['release_year']<2000]
movies_1990s = released_up_to_1999
#Histogram for movie duration distribution in 1990s
plt.hist(movies_1990s['duration'], bins =10)
plt.xlabel("Movie duration")
plt.ylabel("Number of movies")
plt.title("Distribution of duration for movies released in the 1990s")
plt.show()

#The most ferquent movies in the 1990s
duration = int(np.mean(movies_1990['duration']))
print("The most frequent movie duration in the 1990s : "+ str(duration))

#Filter the short action moves released in the 1990s
short_action_movies_1990 = movies_1990s[movies_1990s['genre']=='Action']

#Counting the number of short action movies release in the 1990s
short_movie_count = 0
for index,movies in short_action_movies_1990.iterrows():
    if movies['duration']<90:
        short_movie_count +=1
    else:
        short_movie_count = short_movie_count
print("The number of short action movies released in the 1990s : "+str(short_movie_count))

#Using .sum() to count short action movies
print("The number of short action movies released in the 1990s : " +str((short_action_movies_1990['duration']<90).sum()))
```
